<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# USA mapping dictionary 
country_mapping = {
    'United States of America': 'USA',
    'U.S.A.': 'USA',
    'us': 'USA',
    'IND': 'India',
    'United Kingdom of Great Britain and Northern Ireland': 'UK'
}

# Apply the mapping to the Country column
df['Country'] = df['Country'].replace(country_mapping)
print(sorted(df['Country'].dropna().unique()))
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df['Employment'].unique()
```

```python
# Step 1: Split each entry by ';'
df['Employment_split'] = df['Employment'].str.split(';')

# Step 2: Create binary columns (one-hot encoding for multi-label data)
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
employment_encoded = pd.DataFrame(mlb.fit_transform(df['Employment_split']),
                                  columns=mlb.classes_,
                                  index=df.index)

# Step 3: Combine with original DataFrame (optional)
df_encoded = pd.concat([df, employment_encoded], axis=1)

df_encoded.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Identify missing values per column
missing_counts = df.isnull().sum()

# Sort columns by missing value count in descending order
missing_sorted = missing_counts.sort_values(ascending=False)

# Display columns with the highest number of missing values
print("Columns with missing values (sorted):")
print(missing_sorted)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
# Step 1: Identify numerical columns
num_cols = df.select_dtypes(include=['number']).columns

# Step 2: Impute missing values with mean
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)



print("DataFrame after mean imputation for numerical columns:")
df.head(10)
    
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# 2. Impute missing values in 'RemoteWork' with the most frequent value
most_frequent_RemoteWork = df['RemoteWork'].mode()[0]
df['RemoteWork'].fillna(most_frequent_RemoteWork, inplace=True)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# Min-Max normalization
df['ConvertedCompYearly'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
df['ConvertedCompYearly'].describe()
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
import pandas as pd
import numpy as np
# Step 1: Handle missing values (optional, here we drop them)
df = df.dropna(subset=['ConvertedCompYearly'])

# Step 2: Apply log transformation (use np.log1p to handle zero safely)
df['ConvertedCompYearly'] = np.log1p(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
df['YearsCodePro'].value_counts(dropna=False)
```

```python
# Step 1: Replace textual values and convert to numeric
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': '0',
    'More than 50 years': '51'
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Step 2: Define function to assign experience levels
def classify_experience(years):
    if pd.isnull(years):
        return 'Unknown'
    elif years <= 1:
        return 'Beginner'
    elif years <= 5:
        return 'Intermediate'
    elif years <= 10:
        return 'Advanced'
    else:
        return 'Expert'

# Step 3: Apply function to create new column
df['ExperienceLevel'] = df['YearsCodePro'].apply(classify_experience)

df['ExperienceLevel'].value_counts(dropna=False)
```
