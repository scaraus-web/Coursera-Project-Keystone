### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


Task Explanation: Compare Language Usage vs Interest

You're working with two columns:

LanguageHaveWorkedWith: Languages a respondent has used (e.g., "Python; JavaScript; SQL")

LanguageWantToWorkWith: Languages a respondent wants to use (e.g., "Python; Go; Rust")

Each cell contains a semicolon-separated list of programming languages — not single values. That means you need to split and process them first.

Objective:

Compare how often languages appear in both columns.

Visualize what languages are most popular now vs. desired in the future.

Highlight the overlap (languages that appear in both lists) and differences (languages that appear in only one).

 Why it's meaningful:

This comparison shows:

Which languages are popular in practice (usage)

Which languages are gaining interest (want to learn)

Gaps between actual use and aspiration



```python
# 1. Split the strings into separate entries
# Example: split strings into separate rows
df_have = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
df_want = df['LanguageWantToWorkWith'].dropna().str.split(';').explode()
```

```python
# 2. Get value counts
have_counts = df_have.value_counts()
want_counts = df_want.value_counts()
```

```python
# 3. Join counts into one DataFrame
lang_comparison = pd.DataFrame({
    'HaveUsed': have_counts,
    'WantToUse': want_counts
}).fillna(0)
```

```python
# 4. Visualize
# Grouped bar chart:


lang_comparison_sorted = lang_comparison.sort_values('HaveUsed', ascending=False).head(10)

lang_comparison_sorted.plot(kind='bar', figsize=(12, 6))
plt.title('Top Programming Languages: Used vs Wanted')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
!pip install matplotlib-venn
from matplotlib_venn import venn2

venn2(
    [set(df_have.unique()), set(df_want.unique())],
    set_labels=('Used', 'Wanted')
)
plt.title('Overlap of Used and Wanted Languages')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# Get all unique countries in your data
unique_countries = df['Country'].dropna().unique()

# Base region mapping
base_map = {
    'North America': ['United States', 'Canada', 'Mexico'],
    'South America': ['Brazil', 'Argentina', 'Chile', 'Colombia', 'Peru'],
    'Europe': ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain', 'Netherlands', 'Poland',
               'Sweden', 'Russia', 'Ukraine', 'Switzerland', 'Norway', 'Austria', 'Belgium',
               'Denmark', 'Finland', 'Ireland', 'Portugal', 'Czech Republic', 'Hungary', 'Greece', 'Romania'],
    'Asia': ['India', 'China', 'Japan', 'South Korea', 'Indonesia', 'Pakistan', 'Bangladesh',
             'Vietnam', 'Philippines', 'Thailand', 'Turkey', 'Singapore', 'Malaysia', 'Sri Lanka', 'Nepal'],
    'Africa': ['Nigeria', 'South Africa', 'Egypt', 'Kenya', 'Morocco', 'Algeria', 'Ghana', 'Ethiopia', 'Tanzania'],
    'Oceania': ['Australia', 'New Zealand'],
    'Middle East': ['Iran', 'Iraq', 'Israel', 'Saudi Arabia', 'United Arab Emirates',
                    'Jordan', 'Lebanon', 'Qatar']
}

# Build full region_map with default 'Other'
region_map = {}
for region, countries in base_map.items():
    for country in countries:
        region_map[country] = region

# Fill in any missing countries from your dataset as 'Other'
for country in unique_countries:
    if country not in region_map:
        region_map[country] = 'Other'

# Assign the new 'Region' column
df['Region'] = df['Country'].map(region_map)

```

```python
## Write your code here
 
# 3. Count Remote work type by Region

# Group by Region and Language, then count how many respondents in each region reported using each language:

RemoteWork_counts = df.groupby(['Region', 'RemoteWork']).size().reset_index(name='Count')

# Find Most Popular Remote work types per Region

import seaborn as sns
import matplotlib.pyplot as plt

top_RemoteWork = RemoteWork_counts.groupby('Region').apply(lambda x: x.nlargest(5, 'Count')).reset_index(drop=True)

plt.figure(figsize=(12, 10))
sns.barplot(data=top_RemoteWork, x='Region', y='Count', hue='RemoteWork')
plt.title("Top RemoteWork by Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()





```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='YearsCodePro', y='JobSat', color='royalblue', s=100)

# Labels and title
plt.title('Scatter Plot of Years of experience vs Job satisfaction')
plt.xlabel('years of experience')
plt.ylabel('Job satisfaction')
plt.grid(True)
plt.tight_layout()
plt.show()
```

 Explanation:

Pearson: Measures linear correlation (assumes normal distribution).

Spearman: Measures monotonic correlation (based on ranked values, non-parametric).



Use Pearson when the relationship is linear and data is continuous.

Use Spearman when the data is not normally distributed or contains ordinal variables.



```python
df['JobSat'].head(20)
```

```python
# Pearson Correlation (linear relationship)

from scipy.stats import pearsonr, spearmanr

clean_df = df[['YearsCodePro', 'JobSat']].dropna()

# Handle special cases before conversion
clean_df['YearsCodePro'] = clean_df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})

# Now convert safely
clean_df['YearsCodePro'] = pd.to_numeric(clean_df['YearsCodePro'], errors='coerce')




pearson_corr, pearson_p = pearsonr(clean_df['YearsCodePro'], clean_df['JobSat'])
print(f"Pearson correlation: {pearson_corr:.3f} (p-value: {pearson_p:.3e})")
```

Correlation	0.104	Weak positive linear relationship
p-value	2.937e-69	Statistically significant

    Conclusion: People with more coding experience tend to report slightly higher job satisfaction, but the effect is minimal. However, with a large dataset, even small effects can be statistically significant.

```python
# Spearman Correlation (monotonic relationship)
spearman_corr, spearman_p = spearmanr(clean_df['YearsCodePro'], clean_df['JobSat'])
print(f"Spearman correlation: {spearman_corr:.3f} (p-value: {spearman_p:.3e})")
```

Spearman's ρ	0.119	Weak positive ranked relationship
p-value	3.071e-90	Statistically significant
