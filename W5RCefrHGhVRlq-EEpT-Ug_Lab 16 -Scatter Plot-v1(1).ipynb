### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
import matplotlib.pyplot as plt
import seaborn as sns

age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 20,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70,
}
df['AgeNumeric'] = df['Age'].map(age_map)


# Ensure both columns are numeric
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df['AgeNumeric'] = pd.to_numeric(df['AgeNumeric'], errors='coerce')

# Drop rows with missing CompTotal or Age
df_clean = df.dropna(subset=['JobSatPoints_6', 'ConvertedCompYearly','AgeNumeric'])

# Normalize size (optional: adjust multiplier for better visual scaling)
df_clean['Size'] = df_clean['AgeNumeric'].astype(float) * 15

# Sample DataFrame (replace with your actual df)
# Let's assume 'Age' and 'CompTotal' are numeric
# sns.scatterplot(data=df, x='ConvertedCompYearly', y='JobSatPoints_6', alpha=0.6)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='ConvertedCompYearly', y='JobSatPoints_6', size='Size', sizes=(20, 200), alpha=0.5, legend=False)


import matplotlib.ticker as ticker 
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))



# Add titles and labels
plt.title("Scatter Plot of Compensation vs. job satisfaction and age")
plt.xlabel("Compensation")
plt.ylabel("job satisfaction")
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
df['LanguageHaveWorkedWith'].value_counts()
```

```python
df['JobSatPoints_6'].value_counts()
```

```python
# Drop rows where LanguageHaveWorkedWith is NaN
df_clean = df.dropna(subset=['LanguageHaveWorkedWith']).copy()

# Convert semicolon-separated strings to lists
df_clean['Language'] = df_clean['LanguageHaveWorkedWith'].str.split(';')

# Explode the lists into individual rows and reset index
df_clean = df_clean.explode('Language').reset_index(drop=True)

# Strip whitespace
df_clean['Language'] = df_clean['Language'].str.strip()

# Optional: drop rows with missing JobSatPoints_6 (to prevent plot issues)
df_clean = df_clean.dropna(subset=['JobSatPoints_6'])

# Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_clean, x='Language', y='JobSatPoints_6', alpha=0.6)

plt.title("Scatter Plot of Language vs. Job Satisfaction")
plt.xlabel("Language")
plt.ylabel("Job Satisfaction Score")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by Language and calculate the mean satisfaction
language_summary = df_clean.groupby('Language')['JobSatPoints_6'].mean().sort_values(ascending=False)

# Plot the top 10 languages by satisfaction
top_langs = language_summary.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_langs.values, y=top_langs.index, palette='viridis')

plt.title("Average Job Satisfaction by Language")
plt.xlabel("Average Job Satisfaction")
plt.ylabel("Language")
plt.tight_layout()
plt.show()
```

```python
# Group by Language: get mean satisfaction and count of responses
summary = df_clean.groupby('Language')['JobSatPoints_6'].agg(['mean', 'count']).reset_index()
summary = summary.sort_values('mean', ascending=False)

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=summary,
    x='Language',
    y='mean',
    size='count',      # Bubble size by number of responses
    sizes=(50, 500),   # Adjust bubble size scale
    alpha=0.6,
    color='skyblue'
)

plt.title("Average Job Satisfaction by Language")
plt.xlabel("Language")
plt.ylabel("Average Job Satisfaction")
plt.xticks(rotation=45, ha='right')  # Rotate x labels for readability
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 

# Drop rows where LanguageHaveWorkedWith is NaN
df_clean = df.dropna(subset=['Employment']).copy()

# Convert semicolon-separated strings to lists
df_clean['Employment_clean'] = df_clean['Employment'].str.split(';')

# Explode the lists into individual rows and reset index
df_clean = df_clean.explode('Employment_clean').reset_index(drop=True)

# Strip whitespace
df_clean['Employment_clean'] = df_clean['Employment_clean'].str.strip()

# Optional: drop rows with missing JobSatPoints_6 (to prevent plot issues)
df_clean = df_clean.dropna(subset=['JobSatPoints_6'])
df_clean = df_clean.dropna(subset=['ConvertedCompYearly'])





plt.figure(figsize=(10, 6))

# Scatter plot: color by 'MainBranch' (categorical)
sns.scatterplot(
    data=df_clean,
    x='JobSatPoints_6',
    y='ConvertedCompYearly',
    hue='Employment_clean',   # Categorical variable shown with color
    alpha=0.7
)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.title("Compensation vs. Years of Experience by Employment")
plt.xlabel("Years of Professional Coding")
plt.ylabel("Total Compensation")
plt.legend(title='Employment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.
